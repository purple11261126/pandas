### 按单品类型查看数据

```python
# 需要统计的列
used_cols = ['单品名称', '热量', '脂肪热量', '每日建议摄取量（脂肪）', '胆固醇含量', '糖含量']
data_used = data[used_cols]
data_used.head()
```

```python
# 查看有哪些菜单类型
np.unique(data_used['单品名称'])
# 或者：set(data['菜单类型'])
```

```python
# 查看各个营养成分最高的单品
idxs_max_lis = [data[i].idxmax() for i in used_cols[1:]]  # idxmax() 返回最大值的项对应的 index
item_max = list(map(lambda x: data_used['单品名称'][x], idxs_max_lis))  # 找到对应的单品名称

for i, j, k in zip(used_cols[1:], item_max, idxs_max_lis):
    print('{} 最高的单品：{} 数值为：{}'.format(i, j, data.loc[k, i]))
```

```python
# 查看各个营养成分最低的单品
idxs_min_lis = [data_used[i].idxmin() for i in used_cols[1:]]  # 找到 index
item_min = list(map(lambda x: data_used.loc[x, '单品名称'], idxs_min_lis))  # 根据 index 找到单品
for i, j, k in zip(used_cols[1:], item_min, idxs_min_lis):
    print('{} 最低的单品：{} 数值为：{}'.format(i, j, data.loc[k, i]))
```

### 按菜单类型查看数据

```python
# 菜单类型的单品数目分布
cat_grouped = data.groupby('菜单类型')
print('菜单类型的单品数目：')
print(cat_grouped.size())
```

```python
# 菜单类型的营养成分分布
print('菜单类型的营养成分分布：')

cols_used = ['热量', '脂肪热量', '每日建议摄取量（脂肪）', '胆固醇含量', '糖含量']
print(cat_grouped[cols_used].mean())
```

```python
# 营养成分最高的菜单类型
max_cats = [cat_grouped[col].mean().idxmax() for col in cols_used]
for col, cat in zip(cols_used, max_cats):
    print('{} 最高的菜单类型：{}'.format(col, cat))
```

```python
# 营养成分最低的菜单类型
min_cats = [cat_grouped[col].mean().idxmin() for col in cols_used]
for col, cat in zip(cols_used, min_cats):
    print('{} 最低的菜单类型：{}'.format(col, cat))
```

### 查看分析单品及菜单的份量

```python
data['单品份量'].head()
```

```python
# 过滤数据，只保留包含 'g'的单品
sel_menu_data = data[data['单品份量'].str.contains('g')].copy()  # 找出包含 g 的所有数据

def proc_size_str(size_str):
    """
    处理 serving size 字符串，返回分量的数值
    """
    start_idx = size_str.index('(') + 1
    end_idx = size_str.index('g')
    size_val = size_str[start_idx:end_idx]
    return float(size_val)

sel_menu_data['Size'] = sel_menu_data['单品份量'].apply(proc_size_str)  # 添加分量数值的列
sel_menu_data.head()
```

```python
# 份量最多的单品
max_idx = sel_menu_data['Size'].idxmax()
print('份量最多的单品：{}，{}g'.format(sel_menu_data.iloc[max_idx]['单品名称'], sel_menu_data['Size'].max()))

min_idx = sel_menu_data['Size'].idxmin()
print('份量最少的单品：{}，{}g'.format(sel_menu_data.iloc[min_idx]['单品名称'], sel_menu_data['Size'].min()))
```

```python
sel_cat_grouped = sel_menu_data.groupby('菜单类型')

print('份量最多的类别：{}，{}g'.format(sel_cat_grouped['Size'].mean().idxmax(), 
                             sel_cat_grouped['Size'].mean().max()))

print('份量最少的类别：{}，{}g'.format(sel_cat_grouped['Size'].mean().idxmin(), 
                             sel_cat_grouped['Size'].mean().min()))
```
