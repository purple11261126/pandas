```python
data_file = '../data/menu.csv'
data = pd.read_csv(data_file)
data.head()
```

```python
# 更改列名（中文替换）
columns_name = {'Category':'菜单类型', 
                'Item':'单品名称', 
                'Serving Size':'单品份量', 
                'Calories':'热量', 
                'Calories from Fat':'脂肪热量', 
                'Total Fat':'每日建议摄取量（脂肪）', 
                'Saturated Fat':'饱和脂肪含量', 
                'Saturated Fat (% Daily Value)':'每日建议摄取量（饱和脂肪）', 
                'Trans Fat':'反式脂肪酸', 
                'Cholesterol':'胆固醇含量', 
                'Cholesterol (% Daily Value)':'每日建议摄取量（胆固醇含量）', 
                'Sodium':'钠含量', 
                'Sodium (% Daily Value)':'每日建议摄取量（钠）', 
                'Carbohydrates':'碳水化合物含量', 
                'Carbohydrates (% Daily Value)':'每日建议摄取量（碳水化合物）', 
                'Dietary Fiber':'膳食纤维含量', 
                'Dietary Fiber (% Daily Value)':'每日建议摄取量（ 膳食纤维含量）', 
                'Sugars':'糖含量', 
                'Protein':'蛋白质含量', 
                'Vitamin A (% Daily Value)':'每日建议摄取量（维生素A）', 
                'Vitamin C (% Daily Value)':'每日建议摄取量（维生素C）', 
                'Calcium (% Daily Value)':'每日建议摄取量（钙）', 
                'Iron (% Daily Value)':'每日建议摄取量（铁）'}
data.rename(columns=columns_name, inplace=True)
data.head()
```
