## 散点图

```python
import numpy as np
```

```python
# 直接绘制散点图
x = np.arange(1, 9)
y = x
plt.scatter(x, y)
```

```python
# 改变颜色及大小
x = np.arange(1, 9)
y = x

# 需要给每个坐标设置颜色和大小
colors = ['red'] * (len(x) - 1)  
colors.append('green')
sizes = ([100] * (len(x) // 2)) + ([50] * (len(x) - ((len(x) // 2))))

plt.scatter(x, y, s=sizes, c=colors)
```

```python
# zip() 压缩和 zip(* ) 解压缩
l1 = list(range(1, 6))
l2 = list(range(6, 11))
zip_list = list(zip(l1, l2))  # 压缩
print('zip_list: ', zip_list)
print()
x, y = zip(*zip_list)  # 解压缩
print(x)
print(y)
```

```python
# 添加 图例、x 轴标签、y 轴标签 和 标题

plt.scatter(x=x[:2], y=y[:2], c='red', label='samples 1')  # 给图形添加 label 属性，用于图例
plt.scatter(x=x[2:], y=y[2:], c='blue', label='samples 2')

# 图例，loc=4：右下角，framon：是否显示边框，title：图例标题
plt.legend(loc=4, frameon=False, title='Legend')

# x 轴标签
plt.xlabel('x label')  

# y 轴标签
plt.ylabel('y label')  

# title
plt.title('this Title')  
```

## 线图

```python
linear_data = np.arange(1, 9)
quadratic_data = linear_data ** 2
```

```python
# 默认就是线图
plt.plot(linear_data, '-o', quadratic_data, '-o')
plt.plot([22, 44, 66], '--r')

# 添加坐标轴标签及图例
plt.xlabel('x data')
plt.ylabel('y data')
plt.title('Title')
plt.legend(['legend 1', 'legend 2', 'legend 3'])  # 如果没有给图形添加 label 属性，图例就按先后顺序分配

# 填充两条线中的区域
plt.gca().fill_between(range(len(linear_data)), 
                      linear_data, quadratic_data, 
                      facecolor='green', 
                      alpha=0.25)
```

```python
# 绘制横轴为时间的线图
dates = np.arange('2018-01-01', '2018-01-09', dtype='datetime64[D]')  # 利用 np 生成时间列表

# 绘制图形，此时时间标签的方向还需要调整
plt.plot(dates, linear_data, '-o', 
        dates, quadratic_data, '-o')  

# 时间标签转向 45 度
plt.xticks(rotation='45')  

# 调整边界距离
plt.subplots_adjust(bottom=0.25)

# 有些学术制图，需要 LaTeX 语法
plt.title('Quadratic ($x^2$) vs. Linear ($x$)')
```

## 柱状图

```python
x_vals = list(range(len(linear_data)))

plt.bar(x_vals, linear_data, width=0.3)

# group bar chart
# 同一副图中添加新的柱状图
# 注意，为了不覆盖第一个柱状图，需要对x轴做偏移
x_vals2 = [item + 0.3 for item in x_vals]
plt.bar(x_vals2, quadratic_data, width=0.3)
```

```python
# stack bar chart
x_vals = list(range(len(linear_data)))
plt.bar(x_vals, linear_data, width=0.3)
plt.bar(x_vals, quadratic_data, width=0.3, bottom=linear_data)
```

```python
# 横向柱状图
x_vals = list(range(len(linear_data)))
plt.barh(x_vals, linear_data, height=0.3)
plt.barh(x_vals, quadratic_data, height=0.3, left=linear_data)
```
